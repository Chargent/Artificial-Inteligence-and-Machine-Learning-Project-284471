##### **Checking data integrity**

In this part we check if there are null values in the dataset.

```python
# Looking for NaN values in the dataset

nan_count=trains_df.isnull().sum()
print("NaN count in each column :")
nan_count
```

```python
#write afucntion that identify missing values in the dataset
def missing_values_table(df):
    #total missing values
    mis_val=df.isnull().sum()

    #percentage of missing values
    mis_val_percent=100*df.isnull().sum()/len(df)

    #make a table with the results
    mis_val_table=pd.concat([mis_val, mis_val_percent], axis=1)
    
    #rename the columns
    mis_val_table_ren_columns=mis_val_table.rename(columns={0:'Missing Values', 1:'% of Total Values'})

    #sort the table by percentage of missing descending
    mis_val_table_ren_columns=mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:,1]!=0].sort_values('% of Total Values', ascending=False).round(1)
    
    #return the dataframe with missing information
    return mis_val_table_ren_columns

#run the function to see the missing values
missing_values_table(trains_df)
```

```python
# Filter rows where 'Arrival Delay in Minutes' is NaN
missing_arrival_delay_rows = trains_df[trains_df['Arrival Delay in Minutes'].isnull()]

# Select columns to create distribution plots
selected_columns = ['Age', 'Gender', 'Loyalty', 'Work or Leisure', 'Ticket Class', 
                    'Onboard General Rating', 'Cleanliness Rating', 'Food\'n\'Drink Rating']

# Create a figure with subplots
fig, axes = plt.subplots(len(selected_columns), 1, figsize=(10, 15))

# Plot distribution plots for each selected column
for i, col in enumerate(selected_columns):
    sns.histplot(data=missing_arrival_delay_rows, x=col, ax=axes[i], kde=True, color='red', label='Missing Arrival Delay')
    sns.histplot(data=trains_df, x=col, ax=axes[i], kde=True, color='blue', label='Overall Dataset', alpha=0.5)
    axes[i].set_title(f'Distribution of {col} (Missing "Arrival Delay in Minutes")')
    axes[i].legend()

# Adjust spacing between subplots for better readability
plt.tight_layout()

# Show the plots
plt.show()
```

## **Data Processing**

---

##### **Partitioning the Dataset into Training and Test Set** 
In the following part we divide the dataset into train and test set. The train set is used to train the model,  while the testing segment is utilized to assess how well the model performs with data it hasn't encountered before.

This division is vital to mitigate the risk of overfitting, so to avoid that our model becomes excessively attuned to the training data, impairing its ability to adapt to new, unseen data effectively.
We used the common split ratio 80/20, where 80% of the data is allocated for training purposes and the remaining 20% is set aside for testing.

In this process, it's essential to remove the target variable from the training data. In our case, the column representing `Satisfied` is what we aim to predict.

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into features and target
x = trains.drop('Satisfied', axis=1)
y = trains['Satisfied'].iloc[:,0]

# Splitting the dataset into train and test sets
train, test, train_check, test_check = train_test_split(x, y, test_size=0.2, random_state=69)
```

---

##### **Scaling**

To prepare the data this process adjusts the range of different features to a standard scale, making sure that no single feature dominates the model due to its larger numeric range. In essence, scaling ensures that each feature contributes equally to the model's predictions, preventing bias towards variables simply because they have larger values.

```python
# Identify numerical and categorical columns 
numerical_columns_lst = x.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_columns_lst = x.select_dtypes(include=['object']).columns.tolist()
```

Here we set up a Pipeline that sets up a scaler to standardize the data.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Create transformers for numerical and categorical data
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine transformers into a ColumnTransformer
transformation = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns_lst),
        ('cat', categorical_transformer, categorical_columns_lst)
    ])

# Create a pipeline with the ColumnTransformer
model_pipeline = Pipeline(steps=[('transformation', transformation)])

# Fit the pipeline to train and test set
train_encoded = model_pipeline.fit_transform(train)
test_encoded = model_pipeline.transform(test)
```

---

##### **Partitioning the training set into training and validation set**
Here we partition again the training set so that later we are able to fine-tune the model's hyperparameters and provide an unbiased evaluation of the model fit during the training phase.

The validation set is not used in the actual training process but it can be used to assess the model's performance at various stages of training.

```python
training, validation_test, training_check, validation_test_check = train_test_split(
    train_encoded, train_check, test_size=0.2, random_state=69)
```

## **Modelling**

---

We're dealing with a classification problem because our target variable is categorical, indicating whether customers are satisfied or not. To tackle this we're using four models:

- **Logistic Regression** 
    
    For binary classification problems with relatively simple, linear relationships between features and the target variable is a solid starting point to get information on the dataset, even if usually is not the most accurate model that is our primary goal.

- **K-Nearest Neighbors (KNN)** 
    
    Is an intuitive method that classifies a data point based on how its neighbors are classified. This model assumes that similar things exist in close proximity. In the context of customer satisfaction, it looks at the 'K' most similar customers (based on the features) to predict a customer's satisfaction. It's beneficial for capturing complex patterns without assuming a specific distribution in the data.

- **Decision Trees** 

    Offers a more flexible approach, handling complex patterns by dividing the data using a set of decision rules. This can be especially useful if there are interactions between features that affect customer satisfaction.

- **Random Forests** 

    build upon classification trees, using an ensemble of trees to improve predictive accuracy and control over-fitting. Given the likely complexity and potential noise in customer satisfaction data, the robustness of random forests makes them a strong candidate for improving prediction performance.

Selecting these four models allows us to compare a simple linear method with more complex non-linear methods, ensuring we explore both straightforward and intricate patterns in the data.

---

```python
# Importing score metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
```

In the next phase of our analysis, we will rigorously evaluate each selected model (Logistic Regression, K-Nearest Neighbors (KNN), Decision Trees, and Random Forests) using two key metrics: accuracy score and cross-validation. 

The accuracy score will provide us with a straightforward measure of each model's performance, indicating the proportion of correct predictions made out of all predictions. 

However, to ensure a more robust and unbiased assessment, we will also employ cross-validation. This technique involves dividing the dataset into multiple segments, training the model on some segments and testing it on others, and then averaging the results. 

By using both accuracy score and cross-validation, we can gain comprehensive insights into which model yields the most precise and reliable predictions. This dual-metric approach will enable us to compare the models not just on their immediate predictive power but also on their ability to generalize well across different data samples, ensuring we identify the model that most effectively decodes the intricate patterns within our data.

---

##### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Initializing the model
Logistic_Regression = LogisticRegression(max_iter=1000)

# Training the model
Logistic_Regression.fit(training, training_check)

# Predicting the target values from the validation data
prediction_LR = Logistic_Regression.predict(validation_test)

# Calculating the performance metrics
accuracy_LR = accuracy_score(validation_test_check, prediction_LR)
precision_LR = precision_score(validation_test_check, prediction_LR)
recall_LR = recall_score(validation_test_check, prediction_LR)
f1_LR = f1_score(validation_test_check, prediction_LR)

# Calculting the ROC Score
fpr, tpr, thresholds = roc_curve(validation_test_check, prediction_LR)
auc_LR = auc(fpr, tpr)

# Plotting the ROC Curve
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.rcParams['figure.figsize'] = [4, 4]
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Printing the results in a dataframe
results_lr = pd.DataFrame(data=[['Logistic Regression', accuracy_LR, precision_LR, recall_LR, f1_LR, auc_LR]], 
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])
results_lr
```

```python
cm = confusion_matrix(validation_test_check, prediction_LR)
cm = pd.DataFrame(cm, index=['Actual Negative', 'Actual Positive'], columns=['Predicted Negative', 'Predicted Positive'])
cm
```
